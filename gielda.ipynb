```python

import pandas as pd
import requests
import io
import datetime
```

```python
class StooqReader:
    
    def __init__(self,symbol,start,end):
        self.symbol=symbol
        self.start=start
        self.end=end
        self.url="https://stooq.pl/q/d/l/?s="+self.symbol+"&d1="+self.start+"&d2="+self.end+"&i=d"
    
    def read(self,col='all'):
        req=requests.get(self.url)
        if req.status_code!=200:
            print("Nie znaleziono strony")
            return None
        plik=req.content
        dane=pd.read_csv(io.StringIO(plik.decode('utf-8')))
        if dane.shape[0]==0:
            print("Zły zakres")
            return None
        dane = dane.set_index('Data')
        if col=='all':
            return dane
        else:
            return dane[col]
                
```

```python
lista_wig = pd.read_csv("wig_ListaSpolek.csv")['Symbol']
```

```python
ss=[]

for i in lista_wig:
    print(i)
    x=StooqReader(str(i),'20170101','20180901')
    try:
        x.read(['Zamkniecie','Wolumen'])
        x['delta']=(x['Zamkniecie']-x['Zamkniecie'].shift(1))/x['Zamkniecie'].shift(1)
        ss.append(x['delta'].mean())
    except (TypeError, NameError):
        pass
```
